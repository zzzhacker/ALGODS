<ul>
	<li><code>1 &lt;= arr.length &lt;= 5000</code></li>
	<li><code>0 &lt;= arr[i] &lt;= 9</code></li>
	<li>Each node's value is between [0 - 9].</li>
</ul>

```python
from typing import List,Optional
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
# class Solution:
#     def isValidSequence(self, root: Optional[TreeNode], arr: List[int]) -> bool:
#         ans = [False]
#         def checkPath(root,curr_path):
#             if root is None:
#                 return False
#             curr_path.append(root.val)
#             if root.left is None and root.right is None and curr_path==arr:
#                 if not ans[0]: ans[0] = True
#             else:
#                 checkPath(root.left,curr_path) 
#                 checkPath(root.right,curr_path)
#             curr_path.pop()
#         checkPath(root,[])
#         return ans[0]
class Solution:
    def isValidSequence(self, root: Optional[TreeNode], arr: List[int]) -> bool:
        def checkPath(root,idx):
            if root is None:
                return False
            if idx>len(arr)-1:
                return False
            elif root.left is None and root.right is None and root.val==arr[idx] and idx+1 == len(arr):
                return True
            elif root.val != arr[idx]: return False
            else:  return checkPath(root.left,idx+1) or checkPath(root.right,idx+1)
        
        return checkPath(root,0)
```
